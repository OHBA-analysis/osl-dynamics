```python
sample_length = 20
n_bins = 7
bin_size, remainder = divmod(sample_length, n_bins)

starts = np.arange(1, sample_length + 1, bin_size+1)
ends = starts + bin_size

weights = np.zeros((n_bins, sample_length + 2))

col_indices = np.arange(sample_length)
mask = (starts[:, None] <= col_indices) & (col_indices < ends[:, None])

print(starts)
print(ends)
```

```python
def split_two_bins(interval_start, interval_end, mode="overlap"):
    sample_length = interval_end - interval_start
    n_bins = 2

    if mode == "overlap":
        rounding = np.ceil
    elif mode == "exclude":
        rounding = np.floor
    else:
        raise ValueError("Must be overlap or exclude")
    
    bin_length = rounding(sample_length / n_bins)
    first_bin = np.arange(interval_start, interval_start + bin_length)
    second_bin = np.arange(interval_end - bin_length, interval_end)
    return [first_bin, second_bin]
    
```

```python
split_two_bins(10, 21, mode="exclude")
```
